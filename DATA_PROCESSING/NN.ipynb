```python
import h5py
import numpy as np
import ROOT
```

```python
#Energy Calo
f = h5py.File('../File_En/Outfile_CellInformation_HomDet_2to5GeV_FineGran128_V1.h5','r')
fN= h5py.File('../File_En/Outfile_MLPred_HomDet_2to5GeV_FineGran128_V1.h5','r')
```

```python
f.keys()
```

```python
#h_pl.Delete()
#h_pl1.Delete()


h_pl=ROOT.TH1D("cell" ,"cell",1000,-10,10)#,100,-2,2) 
h_pl1=ROOT.TH1D("event" ,"event",1000,-10,10)#,100,-2,2) 


c1=0
c2=0

layers=[64,32,32,16,16,8]

for ev in range(6000):
    if ev%100==0: print("Ev: ",ev)
    EEvT=0
    EEvNN=0
    
    for l in range(6):
        ET=f["RealRes_NeutralEnergy_Layer"+str(l+1)][ev][0]
        ENN=fN["RealRes_NeutralEnergyPred_Layer"+str(l+1)][ev][0]

        
        for X in range(layers[l]):
            for Y in range(layers[l]):

                if ET[X][Y]!=0.0:
                    F1=(ENN[X][Y]-ET[X][Y])/ET[X][Y]
                    h_pl.Fill( F1 )
                    c1+=1
                    EEvT+=ET[X][Y]

                if ENN[X][Y]!=0.0:
                    EEvNN+=ENN[X][Y]
               
    if EEvT!=0.0:
        c2+=1
        F2=(EEvNN-EEvT)/(EEvT)
        h_pl1.Fill( F2 )

print(c1,c2)
```
